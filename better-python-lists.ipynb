```python
from functools import reduce
import numpy as np
```

```python
class List(list):
    def __init__(self, l: list = list()):
        self.list = list(l)
    
    def __str__(self):
        return str(self.list)
        
    def __repr__(self):
        return f'List({self})'
    
    def __getitem__(self, idx):
        return self.list[idx]
    
    def __setitem__(self, idx, value):
        self.list[idx] = value

    def __iter__(self):
        return iter(self.list)
        
    def __len__(self):
        return len(self.list)
    
    def __contains__(self, item):
        return item in self.list
        
    def map(self, fn):
        return List(map(fn, self.list))
        
    def filter(self, fn):
        return List(filter(fn, self.list))
    
    def reduce(self, fn, *args):
        return reduce(fn, self.list, *args)
    
    def apply(self, fn): # forEach
        self.map(fn)
    
    def length(self):
        return len(self)
    
    def contains(self, item):
        return item in self
    
    def join(self, string):
        return string.join(self.map(str).list)
    
    def sort(self, key=None, reverse=False):
        self.list.sort(key=key, reverse=reverse)
        
    def sorted(self, key=None, reverse=False):
        return List(sorted(self.list, key=key, reverse=reverse))

    def reverse(self):
        self.list.reverse()
        
    def reversed(self):
        return List(reversed(self.list))
    
    def append(self, item):
        self.list.append(item)
        
    def index(self, item):
        return self.list.index(item)
        
    def min(self):
        return min(self.list)
    
    def argmin(self):
        return self.index(self.min())
        
    def max(self):
        return max(self.list)
    
    def argmax(self):
        return self.index(self.max())
        
    def sum(self):
        return sum(self.list)
    
    def mean(self):
        return self.sum() / self.length()
```
