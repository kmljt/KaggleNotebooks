## Feature Engineering

```python
legends['agility'] = legends.strength + legends.dexterity
legends['resilience'] = legends.dexterity + legends.defence
legends['evasion'] = legends.defence + legends.speed
legends['power'] = legends.strength + legends.speed
legends['fortitude'] = legends.strength + legends.defence
legends['nimbleness'] = legends.dexterity + legends.speed
```
