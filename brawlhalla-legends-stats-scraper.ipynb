```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import os
```

```python
response = requests.get('https://brawlhalla.fandom.com/wiki/Brawlhalla_Wiki')
response
```

```python
brawl_wiki = BeautifulSoup(response.content, 'html.parser')
legends = brawl_wiki.select('#content-legends-wrapper > div:nth-child(2) > div:nth-child(2) > div')[0].find_all('p')

legends
```

```python
names = []
stats = []
```

```python
for legend in legends:
    legend_href = 'https://brawlhalla.fandom.com' + legend.contents[0]['href']
#     print(legend_href)
    
    res = requests.get(legend_href)
    legend_soup = BeautifulSoup(res.content, 'html.parser')
    
    name = legend_soup.select('#firstHeading > span')[0].text
    stat = list(map(int, legend_soup.select('#mw-content-text > div > table.wikitable > tbody > tr:nth-child(2) > td:nth-child(1)')[0].text.split()))
    
    names.append(name)
    stats.append(stat)
    
    print(name)
    print(stat)
    print()
```

```python
# adding vivi

names.append('Vivi')
stats.append([6, 5, 4, 7])
```

```python
strength = []
dexterity = []
defence = []
speed = []
```

```python
for stat in stats:
    st, dex, df, sp = stat
    
    strength.append(st)
    dexterity.append(dex)
    defence.append(df)
    speed.append(sp)
```

```python
df = pd.DataFrame({'name': names, 'strength': strength, 'dexterity': dexterity, 'defence': defence, 'speed': speed})
df.set_index('name', inplace=True)

df
```

```python
output_dir = '/kaggle/working'
df.to_csv(os.path.join(output_dir, 'brawlhalla-stats.csv'))
```
