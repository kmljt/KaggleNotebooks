```python
from tqdm.auto import tqdm
import time

pbar = tqdm(total=100, desc='Description')
for i in range(100):
    time.sleep(0.1)
    pbar.update(1)
pbar.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 2.5, 3, 4, 5, 6]
y = [2.4, 14.4, 384, 2000, 100_000, 10_000_000, 100_000_000] # kbps

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_yscale('log')
ax.set_xlabel('Generation (G)')
ax.set_ylabel('Max Speed (kbps)')
ax.set_title('Max Speed for Wireless Communication Technology')
ax.grid()
ax.axhline(1, c='k')
ax.axvline(1, c='k')

plt.show()
```
