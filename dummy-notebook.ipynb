```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 2.5, 3, 4, 5, 6]
y = [2.4, 14.4, 384, 2000, 100_000, 10_000_000, 100_000_000] # kbps

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_yscale('log')
ax.set_xlabel('Generation (G)')
ax.set_ylabel('Max Speed (kbps)')
ax.set_title('Max Speed for Wireless Communication Technology')
ax.grid()
ax.axhline(1, c='k')
ax.axvline(1, c='k')

plt.show()
```

```python
import numpy as np
from matplotlib.ticker import MaxNLocator

train_losses = np.random.rand(10)
val_losses = np.random.rand(10)
r2s = np.random.rand(10)
lrs = [10 ** -i for i in range(10)]
best_r2_epoch = 5
```

```python
plt.figure(figsize=(8, 15))
plt.axis('off')

# ========== (1) ==========

plt.subplot(3, 1, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
if len(train_losses) < 20:
    plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.grid(True)
# plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.axvline(best_r2_epoch, color='cyan')
plt.legend(loc='best')
# plt.ylim(-0.05, 1)

# ========== (2) ==========

plt.subplot(3, 1, 2)
plt.plot(r2s, label='R2 Score')
plt.xlabel('Epoch')
plt.ylabel('Metrics')
if len(train_losses) < 20:
    plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.grid(True)
# plt.axhline(1, color='black')
plt.axvline(0, color='black')
plt.axvline(best_r2_epoch, color='cyan')
plt.legend(loc='best')

# ========== (3) ==========

plt.subplot(3, 1, 3)
plt.yscale('log')
plt.plot(lrs, label='Learning Rate')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
if len(train_losses) < 20:
    plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.axvline(best_r2_epoch, color='cyan')
plt.legend(loc='best')

plt.show()
```
